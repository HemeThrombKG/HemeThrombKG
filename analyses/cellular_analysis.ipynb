### Load HemeThrombKG

```python
review_kg = pybel.from_bel_script('../bel_files/hemeThrombKG.bel')
```

### Load cell terms 

```python
cell_terms_df = pd.read_csv('../node_type_files/cell_terms.tsv', sep='\t', header=None)
cell_terms = cell_terms_df[0].to_list()
```

### Render HemeThrombKG subgraph of cell and cell-related terms

```python
cell_terms_set = get_cellular_node_paths(review_kg, cell_terms)
cell_nodes = get_node_paths(review_kg, cell_terms_set)
```

```python
pybel.to_jupyter(review_kg.subgraph(cell_nodes))
```
